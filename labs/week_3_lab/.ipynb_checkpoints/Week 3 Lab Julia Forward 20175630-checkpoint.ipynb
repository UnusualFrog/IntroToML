# 4) Treat outliers by (a) trimming and (b) winsorising; compare effects.

```python
# Winsorising
df_winsor = df.copy()

def iqr_mask(s, k=1.5):
    q1, q3 = s.quantile(0.25), s.quantile(0.75)
    iqr = q3 - q1
    lower, upper = q1 - k*iqr, q3 + k*iqr
    return (s < lower) | (s > upper), (lower, upper)

df_out = df_winsor.copy()  # start from simply imputed
for c in ['temp_c','wind_kph','precip_mm', 'demand']:
    mask, (lo, hi) = iqr_mask(df_out[c])
    print(f'{c}: {mask.sum()} outliers (IQR fences {lo:.2f}..{hi:.2f})')
    # Winsorize: clip to inner fences
    df_out[c] = df_out[c].clip(lo, hi)

# Check hist after winsorization
for c in ['temp_c','wind_kph','precip_mm', 'demand']:
    plt.figure()
    plt.hist(df_out[c], bins=40)
    plt.title(f'Histogram after winsorization: {c}')
    plt.xlabel(c); plt.ylabel('Count')
    plt.show()
```

```python
# Trimming
df_trim = df.copy()

def iqr_mask(s, k=1.5):
    q1, q3 = s.quantile(0.25), s.quantile(0.75)
    iqr = q3 - q1
    lower, upper = q1 - k*iqr, q3 + k*iqr
    return (s < lower) | (s > upper), (lower, upper)

df_out_trim = df_trim.copy()  # start from simply imputed
for c in ['temp_c','wind_kph','precip_mm', 'demand']:
    mask, (lo, hi) = iqr_mask(df_out_trim[c])
    print(f'{c}: {mask.sum()} outliers (IQR fences {lo:.2f}..{hi:.2f})')
    # Trim
    # Mask contains outliers
    # Use '~' to get mask with only non-outliers
    # Apply flipped mask to get only values which are not outliers
    df_out_trim = df_out_trim[~mask]
    

# Check hist after trimming
for c in ['temp_c','wind_kph','precip_mm', 'demand']:
    plt.figure()
    plt.hist(df_out_trim[c], bins=40)
    plt.title(f'Histogram after trimming: {c}')
    plt.xlabel(c); plt.ylabel('Count')
    plt.show()
```

## 5) Impute missing data: simple, KNN, Iterative, hot-deck by HOUR.

```python
# Median for numeric; mode for categorical

# df_simple_imp = df.copy()
df_simple_imp = df_winsor.copy()
# df_simple_imp = df_trim.copy()

num_cols = df_simple_imp.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = df_simple_imp.select_dtypes(exclude=[np.number]).columns.tolist()

for c in num_cols:
    df_simple_imp[c] = df_simple_imp[c].fillna(df_simple_imp[c].median())

for c in cat_cols:
    df_simple_imp[c] = df_simple_imp[c].fillna(df_simple_imp[c].mode().iloc[0])

print('Simple imputation complete.')
df_simple_imp.isna().sum()
```

# 6) One-hot encode nominal categories; explain why not label-encode.

```python
# Clean categorical data
df_cat = df_simple_imp.copy()

# Normalize text categories
for c in ['condition','user_type','bike_type']:
    df_cat[c] = df_cat[c].astype(str).str.strip()

print('Category levels after cleaning:')
for c in ['condition','user_type','bike_type']:
    print(c, sorted(df_cat[c].unique()))
```

```python
df_enc = df_cat.copy()

# One-hot/dummy for nominal categories
df_enc = pd.get_dummies(df_enc, columns=['condition','user_type', 'bike_type'], drop_first=True)

print('Encoded columns:')
print([c for c in df_enc.columns if c.startswith('condition_') or c.startswith('user_type_') or c.startswith('bike_type_')])
df_enc.head()
```
