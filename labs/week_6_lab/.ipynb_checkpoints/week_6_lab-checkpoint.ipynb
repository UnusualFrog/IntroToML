```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('equipment_tuning_dataset.csv')
df.head(); df.info(); df.describe(include='all')
```

# 1. Exploratory Data Analysis (EDA)
Identify missing values and outliers, visualize distributions, and analyze class balance.

```python
# Identify missing values and outliers
print(df.isna().mean()) # fraction missing per column

plt.figure(figsize=(15, 5))

plt.subplot(1,3,1)
df[[ 'F1', 'F2', 'F4', 'F6']].boxplot()

plt.subplot(1,3,2)
df[[ 'F3', 'F5', 'F9']].boxplot()

plt.tight_layout()
plt.show()
```

```python
# Handle missing values
df_filled = df

df_filled['F3'] = df['F3'].fillna(df['F3'].median())
df_filled['F9'] = df['F9'].fillna(df['F9'].median())

print(df.isna().mean()) # Check that missing values are removed
```

```python
# Visualize distributions
df_numeric = df_filled[["F1", "F2", "F3", "F4", "F5", "F6", "F9", "F12"]]

for feature in df_numeric:
    sns.displot(df_numeric[feature], element="step", stat="density")
```

```python
# analyze class balance
df_cat = df_filled[["F7", "F8", "F10", "F11"]]

# results = []
f7 = []
f8 = []
f10 = []
f11 = []

for feature in df_cat:
    class_count_percent = df_cat[feature].value_counts() / len(df_cat) * 100
    print(str(class_count_percent))

    for class_value, percentage in class_count_percent.items():
        if feature == "F7":
            f7.append({
              "class_value": class_value,
              "percentage": percentage  
            })
        elif feature == "F8":
            f7.append({
              "class_value": class_value,
              "percentage": percentage  
            })
        elif feature == "F10":
            f10.append({
              "class_value": class_value,
              "percentage": percentage  
            })
        else:
            f11.append({
              "class_value": class_value,
              "percentage": percentage  
            })
        
        # results.append({
        #     'feature': feature,
        #     "class_value": class_value,
        #     "percentage": percentage
        # })

    sns.displot(df_cat[feature], element="step", stat="density")

print(f7)
f7_df = pd.DataFrame(f7)
print(f7_df.head(10))

sns.displot(f7_df, x="class_value", element="step", common_norm=False)
```
